```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
njx = pd.read_csv('u05smevosb02_srv01_access_log_gz.csv', header = None, names = ['IP', 'date', 'method', 'url', 'protocol', 'respcode', 'byte'], skipinitialspace=False, sep=';')
```

```python
#find the most loaded day
njx['date'] = njx['date'].str.extract(r'(\d\d/Feb/2015)')
dates = njx.groupby('date')
for date, group in dates:
    print(dates.get_group(date).groupby('date')['protocol'].value_counts())
    #OK it's 9Feb let's read csv one more time
```

```python
njx = pd.read_csv('u05smevosb02_srv01_access_log_gz.csv', header = None, names = ['IP', 'date', 'method', 'url', 'protocol', 'respcode', 'byte'], skipinitialspace=False, sep=';')
```

```python
#drop rows with other date and set format for time
njx['date'] = njx['date'].str.extract(r'09/Feb/2015:(\d\d:\d\d:\d\d)')
njx['date'] = pd.to_datetime(njx['date'], format='%H:%M:%S')
njx.dropna()
```

```python
#drop rows with other date and set format for time
njx['date'] = njx['date'].str.extract(r'09/Feb/2015:(\d\d:\d\d)')
njx['date'] = pd.to_datetime(njx['date'], format='%H:%M')
njx.dropna()
x_dates = []
y_requests = []
minutes = njx.groupby('date')
for date, group in minutes:
    y_requests.append(minutes.get_group(date).groupby('date')['protocol'].value_counts().mean())
    x_dates.append(date)
#njx.groupby('date')['protocol'].value_counts()
```
