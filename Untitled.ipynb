```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
njx = pd.read_csv('u05smevosb02_srv01_access_log_gz.csv', header = None, names = ['IP', 'date', 'method', 'url', 'protocol', 'respcode', 'byte'], skipinitialspace=False, sep=';')
```

```python
#drop rows with other date and set format for time
njx['date'] = njx['date'].str.extract(r'06/Feb/2015:(\d\d:\d\d)')
njx.dropna()
x_dates = []
y_requests = []
minutes = njx.groupby('date')
for date, group in minutes:
    y_requests.append(minutes.get_group(date).groupby('date')['date'].value_counts().mean())
    x_dates.append(date)
```

```python
plt.figure(figsize=(16,8))
plt.xlabel('  :')
plt.ylabel('  ')
plt.title('  ',fontweight="bold")
plt.grid(True)
plt.plot(x_dates,y_requests)

#show every 5 tick only
locs, _ = plt.xticks()
filter_locs, filter_labels = [], []
for ind, (lo, la) in enumerate(zip(locs, x_dates)):
    if ind%5 == 0:
        filter_locs.append(lo)
        filter_labels.append(la)
plt.xticks(filter_locs, filter_labels, rotation=90)
```

```python
requests = njx.groupby('url')
x_dates_3102 = []
y_count_3102 = []
x_dates_3418 = []
y_count_3418 = []
x_dates_3564 = []
y_count_3564 = []
x_dates_3572 = []
y_count_3572 = []
for url, group in requests:
    dates_url = requests.get_group(url).dropna().groupby('date')
    for date, groups in dates_url:
        if url == '/gateway/services/SID0003102/1.00':
            x_dates_3102.append(date)
            y_count_3102.append(dates_url.get_group(date).groupby('date')['url'].value_counts().mean())
        if url == '/gateway/services/SID0003418/1.00':
            x_dates_3418.append(date)
            y_count_3418.append(dates_url.get_group(date).groupby('date')['url'].value_counts().mean())
        if url == '/gateway/services/SID0003564/1.00':
            x_dates_3564.append(date)
            y_count_3564.append(dates_url.get_group(date).groupby('date')['url'].value_counts().mean())
        if url == '/gateway/services/SID0003572/1.00':
            x_dates_3572.append(date)
            y_count_3572.append(dates_url.get_group(date).groupby('date')['url'].value_counts().mean())
```

```python
def paint_graph(url, x_dates, y_counts):
    plt.figure(figsize=(16,8))
    plt.xlabel('  :')
    plt.ylabel('  ')
    plt.title('  '+ url, fontweight="bold")
    plt.grid(True)
    plt.plot(x_dates, y_counts)

    #show every 5 tick only
    locs, _ = plt.xticks()
    filter_locs, filter_labels = [], []
    for ind, (lo, la) in enumerate(zip(locs, x_dates)):
        if ind%5 == 0:
            filter_locs.append(lo)
            filter_labels.append(la)
    plt.xticks(filter_locs, filter_labels, rotation=90)
```

```python
paint_graph('/gateway/services/SID0003102/1.00', x_dates_3102, y_count_3102)
paint_graph('/gateway/services/SID0003418/1.00', x_dates_3418, y_count_3418)
paint_graph('/gateway/services/SID0003564/1.00', x_dates_3102, y_count_3564)
paint_graph('/gateway/services/SID0003572/1.00', x_dates_3572, y_count_3572)
```
