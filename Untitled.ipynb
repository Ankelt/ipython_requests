```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
njx = pd.read_csv('u05smevosb02_srv01_access_log_gz.csv', header = None, names = ['IP', 'date', 'method', 'url', 'protocol', 'respcode', 'byte'], skipinitialspace=False, sep=';')
```

```python
#drop rows with other date and set format for time
njx['date'] = njx['date'].str.extract(r'06/Feb/2015:(\d\d:\d\d)')
njx.dropna()
x_dates = []
y_requests = []
minutes = njx.groupby('date')
for date, group in minutes:
    y_requests.append(minutes.get_group(date).groupby('date')['date'].value_counts().mean())
    x_dates.append(date)
```

```python
plt.figure(figsize=(16,8))
plt.xlabel('Продолжительность теста ч:мм')
plt.ylabel('Запросов в мин')
plt.title('Интенсивность всех запросов',fontweight="bold")
plt.grid(True)
plt.plot(x_dates,y_requests)

#show every 5 tick only
locs, _ = plt.xticks()
filter_locs, filter_labels = [], []
for ind, (lo, la) in enumerate(zip(locs, x_dates)):
    if ind%5 == 0:
        filter_locs.append(lo)
        filter_labels.append(la)
plt.xticks(filter_locs, filter_labels, rotation=90)
```
